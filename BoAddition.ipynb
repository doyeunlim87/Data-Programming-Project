```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import io
```

```python
df = pd.read_csv('COBRA-2009-2018.csv')
dfnparray = pd.DataFrame(df)
dfnparray.dropna()

prob = dfnparray['Neighborhood'].value_counts()
threshold = 4000
mask = prob > threshold
tail_prob = prob.loc[~mask].sum()
prob = prob.loc[mask]
prob.plot(kind='barh')
plt.show()
```
