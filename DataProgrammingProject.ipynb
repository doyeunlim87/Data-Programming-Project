```python
df = pd.read_csv('COBRA-2009-2018.csv')
dfnparray = pd.DataFrame(df)
dfnparray = dfnparray[dfnparray['Occur Date'] > '2008/12/31']
dfnparray = dfnparray[dfnparray['Occur Date'] < '2019/01/01']

```

```python
prob = dfnparray['Neighborhood'].value_counts()
threshold = 4000
mask = prob > threshold
tail_prob = prob.loc[~mask].sum()
prob = prob.loc[mask]
#prob['other'] = tail_prob
prob.plot(kind='barh')
plt.xlabel('Number of Crimes')
plt.show()
```

```python
folium_map1 = folium.Map(location=[33.7557, -84.3884],
                        zoom_start=10.5)
types = dfnparray['UCR Literal'].drop_duplicates()
types = types.reset_index(drop=True)

latdata = dfnparray['Latitude']
londata = dfnparray['Longitude']
typesduplicate = dfnparray['UCR Literal']

latdata = latdata.reset_index(drop=True)
londata = londata.reset_index(drop=True)
typesduplicate = typesduplicate.reset_index(drop=True)

r = 2
w = .1

for i in range(0,len(londata)):
    if typesduplicate[i] == types[0]:
        folium.CircleMarker(location=[latdata[i],londata[i]],radius=r,weight=w,color="blue").add_to(folium_map1)     
    elif typesduplicate[i] == types[1]:
        folium.CircleMarker(location=[latdata[i],londata[i]],radius=r,weight=w,color="green").add_to(folium_map1)
    elif typesduplicate[i] == types[2]:
        folium.CircleMarker(location=[latdata[i],londata[i]],radius=r,weight=w,color="yellow").add_to(folium_map1)
    elif typesduplicate[i] == types[3]:
        folium.CircleMarker(location=[latdata[i],londata[i]],radius=r,weight=w,color="red").add_to(folium_map1)
    elif typesduplicate[i] == types[4]:
        folium.CircleMarker(location=[latdata[i],londata[i]],radius=r,weight=w,color="cyan").add_to(folium_map1)
    elif typesduplicate[i] == types[5]:
        folium.CircleMarker(location=[latdata[i],londata[i]],radius=r,weight=w,color="brown").add_to(folium_map1)
    elif typesduplicate[i] == types[6]:
        folium.CircleMarker(location=[latdata[i],londata[i]],radius=r,weight=w,color="black").add_to(folium_map1)
    elif typesduplicate[i] == types[7]:
        folium.CircleMarker(location=[latdata[i],londata[i]],radius=r,weight=w,color="purple").add_to(folium_map1)
    elif typesduplicate[i] == types[8]:
        folium.CircleMarker(location=[latdata[i],londata[i]],radius=r,weight=w,color="orange").add_to(folium_map1)
    elif typesduplicate[i] == types[9]:
        folium.CircleMarker(location=[latdata[i],londata[i]],radius=r,weight=w,color="olive").add_to(folium_map1)
    elif typesduplicate[i] == types[10]:
        folium.CircleMarker(location=[latdata[i],londata[i]],radius=r,weight=w,color="gray").add_to(folium_map1)

        
folium_map1.save("AllCrimes.html")
```

```python

dates = dfnparray['Occur Date']

dt = [datetime.strptime(str(date),"%Y-%m-%d").date().weekday() for date in dates]







for i in range(0,len(dt)):
    if dt[i] == 0:
        dt[i] = 'Monday'
    elif dt[i] == 1:
        dt[i] = "Tuesday"
    elif dt[i] == 2:
        dt[i] = "Wednesday"
    elif dt[i] == 3:
        dt[i] = "Thursday"
    elif dt[i] == 4:
        dt[i] = "Friday"
    elif dt[i] == 5:
        dt[i] = "Saturday"
    elif dt[i] == 6:
        dt[i] = "Sunday"
        
dtnpa = np.array(dt)
dtDF = pd.DataFrame({'Col1':dtnpa})
dtDF['Col1'].value_counts().plot(kind ='barh')
plt.xlabel('Number of Crimes')
```

```python


dates = dfnparray['Occur Date']

dt2 = [datetime.strptime(str(date),"%Y-%m-%d") for date in dates]

dt2npa = np.array(dt2)
dt2DF = pd.DataFrame({'Col1':dt2npa})
prob = dt2DF['Col1'].value_counts()
prob2 = sorted(prob.index)
fig, ax = plt.subplots(figsize = (10,8))
ax.scatter(prob[prob2].index,prob[prob2].values) 

import matplotlib.dates as mdates
import matplotlib  
x = mdates.date2num(prob[prob2].index)
coeffs = np.polyfit(x,prob[prob2].values,3)
ffit = np.poly1d(coeffs)
p4 = np.poly1d(ffit)
xx = prob[prob2].index
ax.plot(xx, p4(x),'red')
plt.ylabel('Number of Crimes', fontsize=30)
plt.xlabel('Year', fontsize=30)
matplotlib.rc('xtick', labelsize=20)     
matplotlib.rc('ytick', labelsize=20)
print(max(p4(x)))
print(min(p4(x)))
```
